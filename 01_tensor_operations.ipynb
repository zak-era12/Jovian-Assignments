**PyTorch**<br>
The torch package contains data structures for multi-dimensional tensors and mathematical operations over these are defined. Additionally, it provides many utilities for efficient serializing of Tensors and arbitrary types, and other useful utilities. **Tensors**<br>
A tensor is a container which can house data in N dimensions, along with its linear operations, though there is nuance in what tensors technically are and what we refer to as tensors in practice. Often and erroneously used interchangeably with the matrix (which is specifically a 2-dimensional tensor), tensors are generalizations of matrices to N-dimensional space.<br>Mathematically speaking, tensors are more than simply a data container, however. Aside from holding numeric data, tensors also include descriptions of the valid linear transformations between tensors. Examples of such transformations, or relations, include the cross product and the dot product. From a computer science perspective, it can be helpful to think of tensors as being objects in an object-oriented sense, as opposed to simply being a data structure.
While the above is all true, there is nuance in what tensors technically are and what we refer to as tensors as relates to machine learning practice. #### Function 2 - Pytorch range()
PyTorch is an open-source machine learning library developed by Facebook. It is used for deep neural network and natural language processing purposes.

The function torch.range() returns a 1-D tensor of size "((end-start)/step) + 1"
with values from start to end with step step. Step is the gap between two values in the tensor.

 out_{i+1} = out_i + step 

 Syntax: torch.range(start=0, end, step=1, out=None)

 Parameters:
start: the starting value for the set of points. Default: 0.
<br>end: the ending value for the set of points
<br>step: the gap between each pair of adjacent points. Default: 1.
<br>out(Tensor, optional): the output tensor
Return type: A tensor

```python
# Example 1
t5 = torch.range(1, 6) 
t5 
```

This gives us a tensor t5 with the start value as 1 and end value as 6 with the default step value as 1.

```python
# Example 2 
t6=torch.range(-100,0,10) 
t6
```

```python
# Example 3 - breaking (to illustrate when it breaks)
t7=torch.range(100,-100,10)
t7
```

Here it throws a runtime error as the steps are inconsistent with the starting and ending points.

This function can be used to a large tensor just by initialising the start, end and the steps.

#### Function 3 - Pytorch full() function
The function torch.full() returns a tensor of size size filled with fill_value.
Syntax: torch.full(size, fill_value, out=None)

Parameters:
size: a sequence of integers defining the shape of the output tensor
<br>fill_value: the number to fill the output tensor with.
<br>out (Tensor, optional): the output tensor

Return type: A tensor


```python
# Example 1 -
t8=torch.full([5],1.5)
t8
```

This creates a 1D tensor with all 5 elements as 1.5

```python
# Example 2 -
t9=torch.full([5,3,2],-1.5)
t9
```

```python
# Example 3 - breaking (to illustrate when it breaks)
t8=torch.full([2,5],'one')
t8
```

It breaks because the full method accepts only numeric values to be its fill values.

#### Function 4 - torch.mean(input) → Tensor

Returns the mean value of all elements in the input tensor.

Parameters-Parameters
input (Tensor) – the input tensor



```python
# Example 1 - 
t10=torch.randn(3,2)
print(t10)
torch.mean(t10)
```

torch.mean() computes the mean of the randomly generated tensor t10.

```python
# Example 2 - working
t11=torch.randn((2,4,3))
print(t11)
torch.mean(t11, dim=1)
```

Sometimes we may want to calculate mean not on the whole tensor, but on some dimensions of it. In order to do so, we make use of the dim attribute to exclude that from computation of mean.

```python
# Example 3 - breaking (to illustrate when it breaks)
t12= torch.tensor([[11,23],[43,22]])
t12
torch.mean(t12)
```

This is an unexpected behaviour of the torch.mean() method which in normal cases should return the mean of the elements. This can be solved by converting the elements to floating types.

#### Function 5 - torch.trace()
This function returns the sum of the elements of the main diagonal (from the upper left to the lower right) of the input 2-D matrix. The only parameter it takes is an input 2D tensor.

```python
# Example 1 - 
t13=torch.randint(high=10,size=(1,20)).view(4,5)
print(t13)
torch.trace(t13)
```

```python
# Example 2 - 
t14=torch.randn((5,4))
print(t14)
torch.trace(t14)
```

```python
# Example 3 - breaking (to illustrate when it breaks)
t15=torch.randn(3,4,2)
t15
torch.trace(t15)
```

This throws a runtime error as the trace matrix accepts only a 2D matrix and anything else breaks this function.
